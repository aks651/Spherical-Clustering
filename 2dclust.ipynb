```python
#Importing Libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Initialisation
n=100
x_0=2
y_0=1
pi=3.14
r=1
```

```python
angle=np.linspace(0,2*pi,6) 
print(angle)
```

```python
plt.figure()
color=['Green','Blue','Red','Black','Violet']
```

```python
# Plotting 
for i in range(1,len(angle)):
    t=1.25*np.random.uniform(0, 1, n)+angle[i-1]  #Splitting of total angle into five equal parts
    x=x_0+r*np.cos(t)                             # polar cordinate representation
    y=y_0+r*np.sin(t)
    plt.scatter(x,y,s=7,color=color[i-1])       
plt.show()    
```
